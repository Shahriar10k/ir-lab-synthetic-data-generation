```python
# ============================================================
# KoVAE Real vs Synthetic: Realism + Diversity Evaluation
# One-cell notebook version
# No saving by default
# ============================================================

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# ============================================================
# Project paths
# ============================================================

PROJECT_ROOT = Path("/home/iailab42/khans1/projects")

REAL_DIR = PROJECT_ROOT / "data"
SYN_DIR = PROJECT_ROOT / "data" / "synthetic"

REAL_X_PATH = REAL_DIR / "all_X.npy"
REAL_Y_PATH = REAL_DIR / "all_y.npy"
REAL_SUBJECT_PATH = REAL_DIR / "all_subject.npy"
REAL_METADATA_PATH = REAL_DIR / "all_metadata.csv"

# SYN_X_PATH = SYN_DIR / "all_X_synthetic.npy"
# SYN_Y_PATH = SYN_DIR / "all_y_synthetic.npy"
# SYN_SUBJECT_PATH = SYN_DIR / "all_subject_synthetic.npy"
# SYN_METADATA_PATH = SYN_DIR / "all_metadata_synthetic.csv"

SYN_X_PATH = SYN_DIR / "timevae"/ "all_X_synthetic.npy"
SYN_Y_PATH = SYN_DIR / "timevae"/ "all_y_synthetic.npy"
SYN_SUBJECT_PATH = SYN_DIR / "timevae"/ "all_subject_synthetic.npy"
SYN_METADATA_PATH = SYN_DIR / "timevae"/ "all_metadata_synthetic.csv"

OUT_DIR = PROJECT_ROOT / "results" / "kovae"/ "realism_diversity_kovae"
OUT_DIR = PROJECT_ROOT / "results" / "timevae"/ "realism_diversity_timevae"
OUT_DIR.mkdir(parents=True, exist_ok=True)


# ============================================================
# Constants
# ============================================================

TARGET_HZ = 64
WINDOW_LEN = 512
SHIFT_LEN = 128

CHANNEL_NAMES = ["ACC_x", "ACC_y", "ACC_z", "BVP", "EDA", "TEMP"]
NUM_CHANNELS = len(CHANNEL_NAMES)

ACTIVITY_IDS = [1, 2, 3, 4, 5, 6, 7, 8]

REAL_SUBJECT_TO_PLOT = "S1"
SYN_SUBJECT_TO_PLOT = "SYN_001"

MAX_REAL_WINDOWS_PER_ACTIVITY_FOR_DIVERSITY = 1000
MAX_SYN_WINDOWS_PER_ACTIVITY_FOR_DIVERSITY = 1000
DISTANCE_BATCH_SIZE = 256

RANDOM_SEED = 42


# ============================================================
# Basic helpers
# ============================================================

def require_file(path):
    path = Path(path)
    if not path.exists():
        raise FileNotFoundError(f"Missing file: {path}")
    return path


def check_window_shape(X, expected_channels, name):
    if X.ndim != 3:
        raise ValueError(f"{name}: expected [N, T, C], got {X.shape}")
    if X.shape[1] != WINDOW_LEN:
        raise ValueError(f"{name}: expected window length {WINDOW_LEN}, got {X.shape[1]}")
    if X.shape[2] != expected_channels:
        raise ValueError(f"{name}: expected {expected_channels} channels, got {X.shape[2]}")


def safe_mean(x):
    return float(np.mean(x)) if len(x) else np.nan


def safe_std(x):
    return float(np.std(x)) if len(x) else np.nan


def histogram_overlap(real_values, syn_values, bins=80):
    real_values = np.asarray(real_values, dtype=np.float32).reshape(-1)
    syn_values = np.asarray(syn_values, dtype=np.float32).reshape(-1)

    real_values = real_values[np.isfinite(real_values)]
    syn_values = syn_values[np.isfinite(syn_values)]

    if len(real_values) == 0 or len(syn_values) == 0:
        return np.nan

    vmin = float(min(real_values.min(), syn_values.min()))
    vmax = float(max(real_values.max(), syn_values.max()))

    if vmin == vmax:
        return 1.0

    hist_real, bin_edges = np.histogram(real_values, bins=bins, range=(vmin, vmax))
    hist_syn, _ = np.histogram(syn_values, bins=bin_edges)

    hist_real = hist_real.astype(np.float64)
    hist_syn = hist_syn.astype(np.float64)

    hist_real /= hist_real.sum()
    hist_syn /= hist_syn.sum()

    return float(np.minimum(hist_real, hist_syn).sum())


def ks_statistic_sampled(real_values, syn_values, max_points=100000, seed=42):
    rng = np.random.default_rng(seed)

    real_values = np.asarray(real_values, dtype=np.float32).reshape(-1)
    syn_values = np.asarray(syn_values, dtype=np.float32).reshape(-1)

    real_values = real_values[np.isfinite(real_values)]
    syn_values = syn_values[np.isfinite(syn_values)]

    if len(real_values) == 0 or len(syn_values) == 0:
        return np.nan

    if len(real_values) > max_points:
        real_values = real_values[rng.choice(len(real_values), size=max_points, replace=False)]

    if len(syn_values) > max_points:
        syn_values = syn_values[rng.choice(len(syn_values), size=max_points, replace=False)]

    real_sorted = np.sort(real_values)
    syn_sorted = np.sort(syn_values)
    all_values = np.sort(np.concatenate([real_sorted, syn_sorted]))

    real_cdf = np.searchsorted(real_sorted, all_values, side="right") / len(real_sorted)
    syn_cdf = np.searchsorted(syn_sorted, all_values, side="right") / len(syn_sorted)

    return float(np.max(np.abs(real_cdf - syn_cdf)))


def average_log_fft(windows, channel_idx):
    if len(windows) == 0:
        return None

    sig = windows[:, :, channel_idx]
    fft = np.fft.rfft(sig, axis=1)
    mag = np.log1p(np.abs(fft))
    return mag.mean(axis=0).astype(np.float32)


def fft_mae_and_corr(real_windows, syn_windows, channel_idx):
    real_fft = average_log_fft(real_windows, channel_idx)
    syn_fft = average_log_fft(syn_windows, channel_idx)

    if real_fft is None or syn_fft is None:
        return np.nan, np.nan

    mae = float(np.mean(np.abs(real_fft - syn_fft)))

    if np.std(real_fft) < 1e-8 or np.std(syn_fft) < 1e-8:
        corr = np.nan
    else:
        corr = float(np.corrcoef(real_fft, syn_fft)[0, 1])

    return mae, corr


# ============================================================
# Load data
# ============================================================

def load_real_data():
    require_file(REAL_X_PATH)
    require_file(REAL_Y_PATH)
    require_file(REAL_SUBJECT_PATH)
    require_file(REAL_METADATA_PATH)

    X = np.load(REAL_X_PATH).astype(np.float32)
    y = np.load(REAL_Y_PATH).astype(np.int64)
    subjects = np.load(REAL_SUBJECT_PATH, allow_pickle=True).astype(str)
    metadata = pd.read_csv(REAL_METADATA_PATH)

    check_window_shape(X, NUM_CHANNELS, "Real X")

    metadata = metadata.copy()
    metadata["subject"] = subjects
    metadata["activity_label"] = y
    metadata["array_index"] = np.arange(len(y), dtype=np.int64)

    return X, y, subjects, metadata


def load_synthetic_data():
    require_file(SYN_X_PATH)
    require_file(SYN_Y_PATH)
    require_file(SYN_SUBJECT_PATH)
    require_file(SYN_METADATA_PATH)

    X = np.load(SYN_X_PATH).astype(np.float32)
    y = np.load(SYN_Y_PATH).astype(np.int64)
    subjects = np.load(SYN_SUBJECT_PATH, allow_pickle=True).astype(str)
    metadata = pd.read_csv(SYN_METADATA_PATH)

    check_window_shape(X, NUM_CHANNELS, "Synthetic X")

    metadata = metadata.copy()
    metadata["synthetic_subject"] = subjects
    metadata["activity_label"] = y
    metadata["array_index"] = np.arange(len(y), dtype=np.int64)

    return X, y, subjects, metadata


real_X, real_y, real_subjects, real_meta = load_real_data()
syn_X, syn_y, syn_subjects, syn_meta = load_synthetic_data()

print("Real X:", real_X.shape)
print("Synthetic X:", syn_X.shape)
print("\nReal label counts:", dict(zip(*np.unique(real_y, return_counts=True))))
print("Synthetic label counts:", dict(zip(*np.unique(syn_y, return_counts=True))))
print("\nSynthetic subjects:", sorted(np.unique(syn_subjects).tolist())[:20])


# ============================================================
# Realism metrics
# ============================================================

def compute_realism_metrics(real_X, real_y, syn_X, syn_y):
    rows = []

    for act in ACTIVITY_IDS:
        real_act = real_X[real_y == act]
        syn_act = syn_X[syn_y == act]

        print(f"Realism | Activity {act}: real={real_act.shape}, synthetic={syn_act.shape}")

        if len(real_act) == 0 or len(syn_act) == 0:
            continue

        for c, channel_name in enumerate(CHANNEL_NAMES):
            real_values = real_act[:, :, c].reshape(-1)
            syn_values = syn_act[:, :, c].reshape(-1)

            real_diff = np.diff(real_act[:, :, c], axis=1).reshape(-1)
            syn_diff = np.diff(syn_act[:, :, c], axis=1).reshape(-1)

            real_abs_diff = np.abs(real_diff)
            syn_abs_diff = np.abs(syn_diff)

            spike_threshold = np.percentile(real_abs_diff, 99.0)

            fft_mae, fft_corr = fft_mae_and_corr(real_act, syn_act, c)

            rows.append({
                "activity": act,
                "channel": channel_name,
                "real_num_windows": int(len(real_act)),
                "syn_num_windows": int(len(syn_act)),
                "real_mean": safe_mean(real_values),
                "syn_mean": safe_mean(syn_values),
                "abs_mean_diff": abs(safe_mean(real_values) - safe_mean(syn_values)),
                "real_std": safe_std(real_values),
                "syn_std": safe_std(syn_values),
                "abs_std_diff": abs(safe_std(real_values) - safe_std(syn_values)),
                "histogram_overlap_higher_better": histogram_overlap(real_values, syn_values),
                "ks_statistic_lower_better": ks_statistic_sampled(
                    real_values,
                    syn_values,
                    seed=RANDOM_SEED + act + c,
                ),
                "real_abs_derivative_mean": safe_mean(real_abs_diff),
                "syn_abs_derivative_mean": safe_mean(syn_abs_diff),
                "abs_derivative_mean_diff": abs(safe_mean(real_abs_diff) - safe_mean(syn_abs_diff)),
                "real_spike_rate_real_99pct_threshold": float(np.mean(real_abs_diff > spike_threshold)),
                "syn_spike_rate_real_99pct_threshold": float(np.mean(syn_abs_diff > spike_threshold)),
                "fft_logmag_mae_lower_better": fft_mae,
                "fft_logmag_corr_higher_better": fft_corr,
            })

    return pd.DataFrame(rows)


realism_df = compute_realism_metrics(real_X, real_y, syn_X, syn_y)

print("\nRealism metrics preview:")
display(realism_df.head(12))

realism_by_channel = (
    realism_df
    .groupby("channel", as_index=False)
    .mean(numeric_only=True)
)

print("\nAverage realism by channel:")
display(realism_by_channel)


realism_df.to_csv(OUT_DIR / "realism_metrics_by_activity_channel.csv", index=False)
realism_by_channel.to_csv(OUT_DIR / "realism_average_by_channel.csv", index=False)


# ============================================================
# Histogram + FFT plots for selected activities
# ============================================================

def plot_hist_fft_for_activity(act):
    real_act = real_X[real_y == act]
    syn_act = syn_X[syn_y == act]

    if len(real_act) == 0 or len(syn_act) == 0:
        print("Missing real or synthetic windows for activity", act)
        return

    fig, axes = plt.subplots(
        nrows=NUM_CHANNELS,
        ncols=2,
        figsize=(16, 3.0 * NUM_CHANNELS),
    )

    for c, channel_name in enumerate(CHANNEL_NAMES):
        real_values = real_act[:, :, c].reshape(-1)
        syn_values = syn_act[:, :, c].reshape(-1)

        ax = axes[c, 0]
        vmin = min(float(real_values.min()), float(syn_values.min()))
        vmax = max(float(real_values.max()), float(syn_values.max()))

        if vmin == vmax:
            vmin -= 1.0
            vmax += 1.0

        bins = np.linspace(vmin, vmax, 80)

        ax.hist(real_values, bins=bins, density=True, alpha=0.45, label="Real")
        ax.hist(syn_values, bins=bins, density=True, alpha=0.45, label="Synthetic")
        ax.set_title(f"Activity {act} | {channel_name} distribution")
        ax.set_xlabel("Normalized value")
        ax.set_ylabel("Density")
        ax.legend()

        ax = axes[c, 1]
        real_fft = average_log_fft(real_act, c)
        syn_fft = average_log_fft(syn_act, c)
        freqs = np.fft.rfftfreq(WINDOW_LEN, d=1.0 / TARGET_HZ)

        ax.plot(freqs, real_fft, linewidth=1.0, label="Real")
        ax.plot(freqs, syn_fft, linewidth=1.0, label="Synthetic")
        ax.set_title(f"Activity {act} | {channel_name} average log FFT")
        ax.set_xlabel("Frequency (Hz)")
        ax.set_ylabel("log(1 + magnitude)")
        ax.legend()

    plt.tight_layout()
    plt.savefig(OUT_DIR / f"hist_fft_activity_{act}.png", dpi=150)
    plt.show()


for act in [1, 5, 8]:
    plot_hist_fft_for_activity(act)


# ============================================================
# Diversity metrics
# ============================================================

def sample_windows_for_activity(X, y, activity_id, max_windows, rng):
    idx = np.where(y == int(activity_id))[0]

    if len(idx) == 0:
        return np.empty((0, WINDOW_LEN, NUM_CHANNELS), dtype=np.float32)

    if len(idx) > max_windows:
        idx = rng.choice(idx, size=max_windows, replace=False)

    return X[idx].astype(np.float32)


def flatten_channel_windows(windows, channel_idx):
    if len(windows) == 0:
        return np.empty((0, WINDOW_LEN), dtype=np.float32)
    return windows[:, :, channel_idx].reshape(len(windows), WINDOW_LEN)


def nearest_mse_distances(A, B, batch_size=256, exclude_self=False):
    A = np.asarray(A, dtype=np.float32)
    B = np.asarray(B, dtype=np.float32)

    if len(A) == 0 or len(B) == 0:
        return np.array([], dtype=np.float32)

    if exclude_self and len(A) != len(B):
        raise ValueError("exclude_self=True requires A and B to have same length.")

    D = A.shape[1]
    B_T = B.T
    B_norm = np.sum(B * B, axis=1)[None, :]

    nearest = []

    for start in range(0, len(A), batch_size):
        end = min(start + batch_size, len(A))
        A_batch = A[start:end]
        A_norm = np.sum(A_batch * A_batch, axis=1)[:, None]

        dist = (A_norm + B_norm - 2.0 * (A_batch @ B_T)) / float(D)
        dist = np.maximum(dist, 0.0)

        if exclude_self:
            rows = np.arange(end - start)
            cols = np.arange(start, end)
            dist[rows, cols] = np.inf

        nearest.append(np.min(dist, axis=1).astype(np.float32))

    return np.concatenate(nearest, axis=0)


def summarize_dist(values, prefix):
    values = np.asarray(values, dtype=np.float32)
    values = values[np.isfinite(values)]

    if len(values) == 0:
        return {
            f"{prefix}_mean": np.nan,
            f"{prefix}_median": np.nan,
            f"{prefix}_p05": np.nan,
            f"{prefix}_p95": np.nan,
        }

    return {
        f"{prefix}_mean": float(np.mean(values)),
        f"{prefix}_median": float(np.median(values)),
        f"{prefix}_p05": float(np.percentile(values, 5)),
        f"{prefix}_p95": float(np.percentile(values, 95)),
    }


def compute_diversity_metrics(real_X, real_y, syn_X, syn_y):
    rng = np.random.default_rng(RANDOM_SEED)
    rows = []

    for act in ACTIVITY_IDS:
        real_sample = sample_windows_for_activity(
            real_X,
            real_y,
            act,
            MAX_REAL_WINDOWS_PER_ACTIVITY_FOR_DIVERSITY,
            rng,
        )

        syn_sample = sample_windows_for_activity(
            syn_X,
            syn_y,
            act,
            MAX_SYN_WINDOWS_PER_ACTIVITY_FOR_DIVERSITY,
            rng,
        )

        print(f"Diversity | Activity {act}: real={real_sample.shape}, synthetic={syn_sample.shape}")

        if len(real_sample) == 0 or len(syn_sample) == 0:
            continue

        for c, channel_name in enumerate(CHANNEL_NAMES):
            real_flat = flatten_channel_windows(real_sample, c)
            syn_flat = flatten_channel_windows(syn_sample, c)

            syn_to_real = nearest_mse_distances(
                syn_flat,
                real_flat,
                batch_size=DISTANCE_BATCH_SIZE,
                exclude_self=False,
            )

            syn_to_syn = nearest_mse_distances(
                syn_flat,
                syn_flat,
                batch_size=DISTANCE_BATCH_SIZE,
                exclude_self=True,
            )

            real_to_real = nearest_mse_distances(
                real_flat,
                real_flat,
                batch_size=DISTANCE_BATCH_SIZE,
                exclude_self=True,
            )

            row = {
                "activity": act,
                "channel": channel_name,
                "real_sampled_windows": int(len(real_sample)),
                "synthetic_sampled_windows": int(len(syn_sample)),
            }

            row.update(summarize_dist(real_to_real, "real_to_nearest_real_mse"))
            row.update(summarize_dist(syn_to_syn, "syn_to_nearest_syn_mse"))
            row.update(summarize_dist(syn_to_real, "syn_to_nearest_real_mse"))

            real_ref = row["real_to_nearest_real_mse_mean"]
            syn_ref = row["syn_to_nearest_syn_mse_mean"]
            syn_real_ref = row["syn_to_nearest_real_mse_mean"]

            row["syn_nn_vs_real_nn_ratio"] = (
                syn_ref / real_ref if np.isfinite(real_ref) and real_ref > 1e-12 else np.nan
            )

            row["syn_nearest_real_vs_real_nn_ratio"] = (
                syn_real_ref / real_ref if np.isfinite(real_ref) and real_ref > 1e-12 else np.nan
            )

            rows.append(row)

    return pd.DataFrame(rows)


diversity_df = compute_diversity_metrics(real_X, real_y, syn_X, syn_y)

print("\nDiversity metrics preview:")
display(diversity_df.head(12))

diversity_by_channel = (
    diversity_df
    .groupby("channel", as_index=False)
    .mean(numeric_only=True)
)

diversity_by_activity = (
    diversity_df
    .groupby("activity", as_index=False)
    .mean(numeric_only=True)
)

print("\nAverage diversity by channel:")
display(diversity_by_channel)

print("\nAverage diversity by activity:")
display(diversity_by_activity)


diversity_df.to_csv(OUT_DIR / "diversity_metrics_by_activity_channel.csv", index=False)
diversity_by_channel.to_csv(OUT_DIR / "diversity_average_by_channel.csv", index=False)
diversity_by_activity.to_csv(OUT_DIR / "diversity_average_by_activity.csv", index=False)


# ============================================================
# Diversity summary plots
# ============================================================

def plot_diversity_summary_by_channel(diversity_df):
    avg = (
        diversity_df
        .groupby("channel", as_index=False)[
            [
                "real_to_nearest_real_mse_mean",
                "syn_to_nearest_syn_mse_mean",
                "syn_to_nearest_real_mse_mean",
            ]
        ]
        .mean()
    )

    x = avg["channel"].astype(str).to_numpy()
    pos = np.arange(len(x))
    width = 0.25

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(pos - width, avg["real_to_nearest_real_mse_mean"], width, label="Real → nearest real")
    ax.bar(pos, avg["syn_to_nearest_syn_mse_mean"], width, label="Synthetic → nearest synthetic")
    ax.bar(pos + width, avg["syn_to_nearest_real_mse_mean"], width, label="Synthetic → nearest real")

    ax.set_xticks(pos)
    ax.set_xticklabels(x)
    ax.set_xlabel("Channel")
    ax.set_ylabel("Nearest-window MSE")
    ax.set_title("Diversity comparison by channel")
    ax.legend()

    plt.tight_layout()
    plt.savefig(OUT_DIR / "diversity_summary_by_channel.png", dpi=150)
    plt.show()


def plot_diversity_summary_by_activity(diversity_df):
    avg = (
        diversity_df
        .groupby("activity", as_index=False)[
            [
                "real_to_nearest_real_mse_mean",
                "syn_to_nearest_syn_mse_mean",
                "syn_to_nearest_real_mse_mean",
            ]
        ]
        .mean()
    )

    x = avg["activity"].astype(str).to_numpy()
    pos = np.arange(len(x))
    width = 0.25

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(pos - width, avg["real_to_nearest_real_mse_mean"], width, label="Real → nearest real")
    ax.bar(pos, avg["syn_to_nearest_syn_mse_mean"], width, label="Synthetic → nearest synthetic")
    ax.bar(pos + width, avg["syn_to_nearest_real_mse_mean"], width, label="Synthetic → nearest real")

    ax.set_xticks(pos)
    ax.set_xticklabels(x)
    ax.set_xlabel("Activity")
    ax.set_ylabel("Nearest-window MSE")
    ax.set_title("Diversity comparison by activity")
    ax.legend()

    plt.tight_layout()
    plt.savefig(OUT_DIR / "diversity_summary_by_activity.png", dpi=150)
    plt.show()


plot_diversity_summary_by_channel(diversity_df)
plot_diversity_summary_by_activity(diversity_df)

print("\nDone.")
```
